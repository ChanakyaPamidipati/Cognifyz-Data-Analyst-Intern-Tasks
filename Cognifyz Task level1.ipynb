# Task 1 : Top cuisines
* <b> Determine the top three most common cuisines in dataset

```python
#Count the occurance of cuisines using => df.value_counts()
#Reset the index of DataFrame using    => reset_index()
value_counts = cuisines.value_counts().reset_index().head(3)
value_counts
```

```python
plt.figure(figsize=(10,5))
values = value_counts['Cuisines']
labels = value_counts['count']
plt.bar(values,labels, color=['red', 'blue', 'green'])
plt.title('Top three Cuisines')
plt.xlabel('Cuisines')
plt.ylabel("Count - Frequency")
plt.show()
```

* <b> Calculate the percentage of restaurants that serve each of the top cuisines

```python

# adding new column 'Percentage' and calculate the percentage of restaurants that serve each of the top cuisines.
value_counts['Percentage'] = round((value_counts['count'] / len(restaurent_df))*100,2)
value_counts
```

```python
plt.title('Percentage of restaurantrants that serve each of the top cuisines.')
plt.pie(value_counts['Percentage'],labels=value_counts['Cuisines'],autopct='%0.2f%%',startangle=90)
plt.show()
```

# Task 2: City Analysis
* <b> Identify the city with the highest number of restaurants in the dataset.

```python
city_restaurant_count = restaurent_df.groupby('City')['Restaurant Name'].count()
city_restaurant_count
```

```python
# Find the city with the highest number of restaurants
max_restaurant_city = city_restaurant_count.idxmax()
max_restaurant_count = city_restaurant_count.max()
print(f"{max_restaurant_city} has highest number of restaurants, the count of restaurants is {max_restaurant_count} restaurants.")
```

* <b>  Calculate the average rating for restaurants in each city. 

```python
avg_ratings_by_city = restaurent_df.groupby("City")["Aggregate rating"].mean().reset_index()
avg_ratings_by_city
```

* <b> Determine the city with the highest average rating

```python
#sort_values,ascending=True => used to sort the values in ascending order
highest_average_rating=avg_ratings_by_city.sort_values(by="Aggregate rating",ascending=True)
highest_average_rating.tail(1)         #tail(1) => Last record
```

##### Observations
* City with the highest average rating
* Inner City with <b>4.9</b> as Average Rating

* <b> Calculate the percentage of restaurants in each price range category.

```python
value_counts = restaurent_df["Price range"].value_counts().reset_index()

# Rename the columns
value_counts.columns = ['Price-Range', 'Count']    
total_count = value_counts['Count'].sum()

value_counts['Percentage'] = round((value_counts['Count'] / total_count)*100,2)

# Print the result as a table
df = pd.DataFrame(value_counts)
df
```

##### Observations

* Percentage of restaurants in each price range category.
    * Price Range of 1 Percentage : 46.53%
    * Price Range of 2 Percentage: 32.59%
    * Price Range of 3 Percentage: 14.74%
    * Price Range of 4 Percentage: 6.14%

# Task 4: Online Delivery

* <b> Determine the percentage of restaurants that offer online delivery

```python
total_restaurant_count =restaurent_df.shape[0]
total_restaurant_count
```

```python
online_restaurnat_count = restaurent_df[restaurent_df['Has Online delivery']=='Yes']
online_restaurant_count= online_restaurnat_count.shape[0]
online_restaurant_count
```

* <b>Compare the average ratings of restaurants with and without online delivery.

```python
restaurent_df["Has Online delivery"]=restaurent_df['Has Online delivery'].str.lower()
with_online = restaurent_df[restaurent_df['Has Online delivery']=='yes']
without_online = restaurent_df[restaurent_df['Has Online delivery']=='no']


avg_of_with_online=with_online['Aggregate rating'].mean()
avg_of_without_online=without_online['Aggregate rating'].mean()
print("Average rating of a restaurant with online delivery:",avg_of_with_online)
print("Average rating of a restaurant without online delivery",avg_of_without_online)
```
